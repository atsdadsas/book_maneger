```python
# 線形探索
# 線形探索の準備
data=[50,30,90,10,20,70,60,40,80]
found =False
def liner_search(data,value):
    for i in range(len(data)):
        if data[i]==value:
            return i
    return -1

print(liner_search(data,40))
```

```python
# 二分探索
def binary_search(data,value):
    left=0
    right=len(data)-1
    while left <=right:
        mid =(left+right)//2
        if data[mid]==value:
            return mid
        elif data[mid]<value:
            left =mid +1
        else:
            right=mid-1
    return -1
    
data=[10,20,30,40,50,60,70,80,90]
print(binary_search(data,90))
```

```python
# 深さ優先探索
tree =[[1,2],[3,4],[],[],[]]
def search(pos):
    print(pos,end="")
    # iに[]が代入されたとき、forはループせずに終了する
    for i in tree[pos]:
        search(i)
search(0)
```

```python
# 通りがけ順の深さ優先探索
tree =[[1,2],[3,4],[5,6],[],[],[],[]]
def search(pos):
    if len(tree[pos])==2:
        search(tree[pos][0])
        print(pos,end="")
        search(tree[pos][1])
    elif len(tree[pos])==1:
        search(tree[pos][0])
        print(pos,end="")
    else:
        print(pos,end="")
search(0)
        
```
