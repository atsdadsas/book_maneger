```python
# h↓　w→
h,w,n=map(int,input().split())
# h*w分の入力。1行目からの入力。それはh個入力される
b=[]
for i in range(h):
    a=list(map(int,input().split()))
    b.append(a)
    
array=[0]
# 行の数＝w
# ; w行分のデータを入力する
ans=[0]*h# ; 司会者の宣言をすべて入れる。

for i in range(n):
    array.append(int(input()))
    
    
# ; 縦の数分ループ
for i in range(h):
# ; 横の数分ループ
    for j in range(w):
# ; arrayに入れられたデータが司会者宣言の中にあるか
        if b[i][j] in array:
            ans[i]+=1
print(max(ans))
            
        
```
