問題１　消費税計算
<br>価格priceを受け取り、消費税10%を加えた税込価格を返すwith_tax(price)関数を作る。少数は四捨五入して整数に。


```python
def with_tax(price):
    price=price*1.1
    return price
x=int(input())
print(int(with_tax(x)))
```

問題２　合格判定
<br>テストの点数scoreを受け取り、60点以上なら「合格」、それ以外は「不合格」を文字列で返す関数judge(score)を作る。

```python
score=int(input())
def  judge(score):
    if score>=60:
        print("合格")
    else:
        print("不合格")

judge(score)
```

問題３　占い
<br>fortune(name)関数を作り、nameを受け取り、"大吉" "中吉" "小吉" "凶"のいずれかをランダムに返す。表示は「〇〇さんの今日の運勢は大吉です！」のように。
ヒント：import random、random.choice()。

```python
import random
name=str(input())
def fortune(name):
    ans=random.choice(["大吉","中吉","小吉"])
    return ans
print(f"{name}さんの今日の運勢は{fortune(name)}です")
```
